```python
# Import liabraries 
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot  as plt
```

```python
# Import dataset
kashti = sns.load_dataset("titanic")
kashti
```

```python
sns.pairplot(kashti)
```

```python
kashti.info()
```

```python
# Check null values befor plotting
kashti.isnull()
```

```python
# Check null values befor plotting
kashti.isnull().sum()
# After checking null values. We will deal carefully with null value
```

```python
# Code for plotting
# Bar plot
kashti.survived.value_counts().plot(kind='bar')
```

```python
sns.countplot(x='survived', data=kashti)
```

```python
sns.countplot(x='age',hue='survived', data=kashti)
```

```python
sns.countplot(x='class', hue= 'sex', data=kashti)
```

```python
# Scatter plot
plt.scatter(x='age',y='fare', data=kashti)
```

```python
sns.scatterplot(x='age',y='fare', data=kashti, hue="class")
```
